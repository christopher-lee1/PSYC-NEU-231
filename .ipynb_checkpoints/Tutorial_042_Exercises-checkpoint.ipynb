# In class exercises for ML tutorials

## Read in data from eeg study
* This time I've kept in data across a 1s window from many electrodes (20) sampled at 250Hz, so now we have a [960, 20, 250] matrix that contains data from all 960 trials, 20 electrodes, and over a 1s window following stimulus onset
* Also read in a "cond_labels" vector that has 960 0's and 1's to mark trials from the two experimental conditions 

## import stuff...

```python
import numpy as np
from scipy.linalg import eigh
from scipy.linalg import sqrtm
import os
from sklearn.svm import SVC 
import matplotlib.pyplot as plt
```

```python
# load the data...
data_path = os.getcwd() + os.sep + 'class_demo_eeg.npz'
eeg = np.load(data_path)

# get the different arrays like this...
eeg['data']
eeg['sr']
eeg['tx']
eeg['cond_labels']

# shape of data
print('Shape of the big eeg data set: ', eeg['data'].shape)

# and if you want to save some typing, especially because we only have a few variables, you reassign the different arrays like this
data = eeg['data']
sr = eeg['sr']
tx = eeg['tx']
cond_labels = eeg['cond_labels']
```

## Compute the mean across the first dimension (trials) and then plot the mean response across time in each of the 20 electrodes on one axis

* Might need to transpose the data into a Time x Electrode format

```python
mean_data = np.mean(data, axis = 0)

plt.figure(figsize=(30,15))
plt.plot(tx, mean_data.T)
plt.xlabel('Time (ms)', {'size':24})
plt.ylabel('Amplitude', {'size':24})
plt.title('Mean EEG Data', {'size':24})
plt.show()
```

## That gives you an idea about what the pattern of data looks like across electrodes. 
* Next, take the mean across time, so you have a 960 x 20 matrix 
* Use the cond_labels vector to sort the data based on condition (so you have two 480 x 20 matrices)
* Do classification to see if you can decode the experimental condition
* Remember - break up your data into independent training and testing sets
* Set C to a small number if this is taking a long time...

```python
mean_data_time = np.mean(data, axis = 2)
print('Mean Data dimensions = ', np.shape(mean_data_time))

cond0 = mean_data_time[cond_labels==0,:]
cond1 = mean_data_time[cond_labels==1,:]

r0,c0 = cond0.shape
r1,c1 = cond1.shape

train_length = int(np.round(.9*r0, 0))

train_stack = np.vstack((cond0[0:train_length,:],cond1[0:train_length,:]))
test_stack = np.vstack((cond0[train_length:,:], cond1[train_length:,:]))

train_label = np.hstack((np.zeros(train_length),np.ones(train_length)))
test_label = np.hstack((np.zeros(r0-train_length),np.ones(r1-train_length)))

mean0 = np.mean(train_stack[train_label==0,:], axis = 1)
mean1 = np.mean(train_stack[train_label==1,:], axis = 1)

model = SVC(kernel='linear', C=2)
model.fit(train_stack, train_label)

class_labels = model.predict(test_stack)

class_accuracy = np.sum(class_labels==test_label) /  test_stack.shape[0]
print('Classification accuracy with a SVM is: ', class_accuracy)
```

## Now try doing this across timepoints!
* First do a time window based on where you think there is nice separation in the responses across electrodes (look at the data that you plotted above)...maybe try 300-400msec post stim
* Then you can try timepoint x timepoint or a moving window

```python
time1 = 300
time2 = 400
data_subset = data[:,:,((tx>=time1) & (tx<time2))]
print('shape of data_subset: ', data_subset.shape)
data_subset_mean = data_subset.mean(axis = 2)

cond0 = data_subset_mean[cond_labels==0,:]
cond1 = data_subset_mean[cond_labels==1,:]

r0,c0 = cond0.shape
r1,c1 = cond1.shape

train_length = int(np.round(.9*r0, 0))

train_stack = np.vstack((cond0[0:train_length,:],cond1[0:train_length,:]))
test_stack = np.vstack((cond0[train_length:,:], cond1[train_length:,:]))

train_label = np.hstack((np.zeros(train_length),np.ones(train_length)))
test_label = np.hstack((np.zeros(r0-train_length),np.ones(r1-train_length)))

mean0 = np.mean(train_stack[train_label==0,:], axis = 1)
mean1 = np.mean(train_stack[train_label==1,:], axis = 1)

model = SVC(kernel='linear', C=2)
model.fit(train_stack, train_label)

class_labels = model.predict(test_stack)

class_accuracy = np.sum(class_labels==test_label) /  test_stack.shape[0]
print('Classification accuracy with a SVM is: ', class_accuracy)
```

```python
time_start = np.arange(0,900,100)

accuracy_array = []

for i in list(time_start):
    data_subset = data[:,:,((tx>=i) & (tx<i+200))]
    data_subset_mean = data_subset.mean(axis = 2)

    cond0 = data_subset_mean[cond_labels==0,:]
    cond1 = data_subset_mean[cond_labels==1,:]

    r0,c0 = cond0.shape
    r1,c1 = cond1.shape

    train_length = int(np.round(.9*r0, 0))

    train_stack = np.vstack((cond0[0:train_length,:],cond1[0:train_length,:]))
    test_stack = np.vstack((cond0[train_length:,:], cond1[train_length:,:]))

    train_label = np.hstack((np.zeros(train_length),np.ones(train_length)))
    test_label = np.hstack((np.zeros(r0-train_length),np.ones(r1-train_length)))

    mean0 = np.mean(train_stack[train_label==0,:], axis = 1)
    mean1 = np.mean(train_stack[train_label==1,:], axis = 1)

    model = SVC(kernel='linear', C=1)
    model.fit(train_stack, train_label)

    class_labels = model.predict(test_stack)

    class_acc = np.sum(class_labels==test_label) /  test_stack.shape[0]
    print('Classification accuracy between times ', i, 'ms and ', i+200, 'ms with a SVM is: ', class_acc)
    
    accuracy_array.append(class_acc)

plt.bar(time_start, accuracy_array)
```
