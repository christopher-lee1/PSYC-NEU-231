```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

### Add percentage tip column to dataset

```python
tips['tip_percent'] = (tips['tip'])/(tips['total_bill']) #assuming total bill does not include tip
tips.head()
```

```python
sns.boxplot(x='time', y='tip_percent', data=tips, hue='day')
```

```python
sns.boxplot(x='size', y='tip_percent', data=tips, hue='day')
plt.title('Lunch and Dinner combined')
plt.show()
sns.boxplot(x='size', y='tip_percent', data = tips[tips['time']=='Lunch'], hue='day')
plt.title('Lunch time')
plt.show()
sns.boxplot(x='size', y='tip_percent', data = tips[tips['time']=='Dinner'], hue='day')
plt.title('Dinner time')
plt.show()
```

2) Find an interesting trend in the "exercise" dataset

```python
sns.set(rc={'figure.figsize':(20,12)})

ax = sns.violinplot(x = "size", y = "tip_percent", data = tips, alpha = 0.2)
ax = sns.stripplot(x = "size", y = "tip_percent", data = tips, jitter = True, color = '0.3')

print("Tip percentage seems to be pretty consistent regardless of party size (except when you go by yourself).")
```
